### Producer Mode -> Writes Data to Broker

````python
from kafka import KafkaProducer
import random
import time
from datetime import datetime

def generate_log_entry():
    """Genera una entrada de log en el formato exacto que espera el consumer"""
    user_id = f"user_{random.randint(1, 10000)}"
    timestamp = int(time.time())
    request_type = "recommendation request"
    status = random.choice([200, 400, 404, 500])
    latency = round(random.uniform(10, 500), 2)
    
    # Formato exacto que espera el consumer:
    # user_id,timestamp,request_type,status=XXX, YYY.YYms
    return f"{user_id},{timestamp},{request_type},status={status}, {latency}ms"

def main():
    producer = KafkaProducer(
        bootstrap_servers=["localhost:9092"],
        value_serializer=lambda x: x.encode('utf-8')  # Simple string encoder
    )
    
    topic = 'lab07'  # Cambia por tu topic real
    
    try:
        while True:
            log_entry = generate_log_entry()
            producer.send(topic, value=log_entry)
            print(f"{datetime.now()} - Sent: {log_entry}")
            
            # Espera aleatoria entre 0.1 y 1 segundo
            time.sleep(random.uniform(0.1, 1))
            
    except KeyboardInterrupt:
        print("\nDeteniendo el producer...")
    finally:
        producer.flush()
        producer.close()

if __name__ == "__main__":
    main()
````
